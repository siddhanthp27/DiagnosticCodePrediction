# Preprocessing Data

```python
import pandas as pd
import numpy as np
```

```python
# load dataframes

df1 = pd.read_csv('DIAGNOSES_ICD.csv')
df2 = pd.read_csv('PROCEDURES_ICD.csv')
```

Store in a dict of dicts
First key of the dict is the patient id.
Each value of the patient id is a dict.
The key of this dict is the hospital admit ID.
The value is the list containing the sequence of the Diagnosis ICD9 Codes.

```python
# print(df1.iloc[0])
# for i in range(len(df2)):
#     if df2.iloc[i]['SUBJECT_ID'] == 109:
#         print(df2.iloc[i])

diag_dict = dict()

for i in range(len(df2)):
    if df2.iloc[i]['SUBJECT_ID'] not in diag_dict.keys():
        diag_dict[df2.iloc[i]['SUBJECT_ID']] = dict()
        if (df2.iloc[i]['HADM_ID'] not in diag_dict[df2.iloc[i]['SUBJECT_ID']].keys()):
            diag_dict[df2.iloc[i]['SUBJECT_ID']][df2.iloc[i]['HADM_ID']] = list()
            diag_dict[df2.iloc[i]['SUBJECT_ID']][df2.iloc[i]['HADM_ID']].append(df2.iloc[i]['ICD9_CODE'])
        else:
            diag_dict[df2.iloc[i]['SUBJECT_ID']][df2.iloc[i]['HADM_ID']].append(df2.iloc[i]['ICD9_CODE'])
    else:
        if (df2.iloc[i]['HADM_ID'] not in diag_dict[df2.iloc[i]['SUBJECT_ID']].keys()):
            diag_dict[df2.iloc[i]['SUBJECT_ID']][df2.iloc[i]['HADM_ID']] = list()
            diag_dict[df2.iloc[i]['SUBJECT_ID']][df2.iloc[i]['HADM_ID']].append(df2.iloc[i]['ICD9_CODE'])
        else:
            diag_dict[df2.iloc[i]['SUBJECT_ID']][df2.iloc[i]['HADM_ID']].append(df2.iloc[i]['ICD9_CODE'])


```

```python
proc_dict = diag_dict
```

The procedure dict is stored in the same structure as above. Instead of using Diagnosis ICD9 Codes, we make use of the Procedure ICD9 codes.

```python
diag_dict = dict()

for i in range(len(df1)):
    if df1.iloc[i]['SUBJECT_ID'] not in diag_dict.keys():
        diag_dict[df1.iloc[i]['SUBJECT_ID']] = dict()
        if (df1.iloc[i]['HADM_ID'] not in diag_dict[df1.iloc[i]['SUBJECT_ID']].keys()):
            diag_dict[df1.iloc[i]['SUBJECT_ID']][df1.iloc[i]['HADM_ID']] = list()
            diag_dict[df1.iloc[i]['SUBJECT_ID']][df1.iloc[i]['HADM_ID']].append(df1.iloc[i]['ICD9_CODE'])
        else:
            diag_dict[df1.iloc[i]['SUBJECT_ID']][df1.iloc[i]['HADM_ID']].append(df1.iloc[i]['ICD9_CODE'])
    else:
        if (df1.iloc[i]['HADM_ID'] not in diag_dict[df1.iloc[i]['SUBJECT_ID']].keys()):
            diag_dict[df1.iloc[i]['SUBJECT_ID']][df1.iloc[i]['HADM_ID']] = list()
            diag_dict[df1.iloc[i]['SUBJECT_ID']][df1.iloc[i]['HADM_ID']].append(df1.iloc[i]['ICD9_CODE'])
        else:
            diag_dict[df1.iloc[i]['SUBJECT_ID']][df1.iloc[i]['HADM_ID']].append(df1.iloc[i]['ICD9_CODE'])
```

Save the two dicts: diag_dict and proc_dict in a pkl file for easy access

```python
diag_pkl = open('diag_dict.pkl', 'rb')
diag_dict = pickle.load(diag_pkl)
diag_pkl.close()

proc_pkl= open('proc_dict', 'rb')
proc_dict = pickle.load(proc_pkl)
proc_pkl.close()
```

Now we shall generate the patient id list over which we will iterate

```python
pat_ids = list()

for key in proc_dict.keys():
    if key in diag_dict.keys():
        pat_ids.append(key)
```

```python
len(pat_ids)
```

```python
train_data = dict()

for i in pat_ids:
    train_data[i] = dict()
    for key in proc_dict[i].keys():
        if key in diag_dict[i].keys():
            train_data[i][key] = list()
            train_data[i][key].append(diag_dict[i][key])
            train_data[i][key].append(proc_dict[i][key])
```

```python
ctr = 0
for key in train_data.keys():
    ctr += len(train_data[key].keys())
```
